```python
PROJECT_ID = 'bdcc-project1-417811'

from google.colab import auth
auth.authenticate_user()
!gcloud config set project {PROJECT_ID}

import google.cloud.bigquery as bq

client = bq.Client(project=PROJECT_ID)

mega_string = ""
```

# CSV creation



```python
def function(description):
  query = '''
            Select ImageId
            FROM bdcc-project1-417811.openimages.image_labels
            JOIN bdcc-project1-417811.openimages.classes USING(Label)
            WHERE Description = @description
            LIMIT 100
          '''

  job_config = bq.QueryJobConfig(
          query_parameters=[
              bq.ScalarQueryParameter("description", "STRING", description)
          ]
      )

  image_search_results = client.query(query, job_config=job_config).result()

  results = image_search_results.to_dataframe()


  image_ids = results['ImageId']
  # print(results)

  i = 0
  baby_string = ""

  for id in image_ids:
    if i < 80:
      baby_string += "training,gs://bdcc-project1-417811-images/" + id + ".jpg," + description + "\n"
    elif i < 90:
      baby_string += "validation,gs://bdcc-project1-417811-images/" + id + ".jpg," + description + "\n"
    else:
      baby_string += "test,gs://bdcc-project1-417811-images/" + id + ".jpg," + description + "\n"

    !gsutil cp gs://bdcc_open_images_dataset/images/{id}.jpg gs://bdcc-project1-417811-images/{id}.jpg
    i += 1

  return baby_string
```

```python
labels = ["Saxophone", "Apple", "Beetle", "Boot", "Coffee", "Cowboy hat", "Cucumber", "Football helmet", "Goat", "Hamster"]

for label in labels:
  mega_string += function(label)

file_path = "/dataset.csv"

# Write the mega_string to the file
with open(file_path, "w") as file:
    file.write(mega_string)

!gsutil cp /dataset.csv gs://bdcc-project1-417811-images/
```
